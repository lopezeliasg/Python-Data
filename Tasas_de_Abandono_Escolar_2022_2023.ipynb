```python
import pandas as pd

url = "https://github.com/lopezeliasg/Python-Data/raw/refs/heads/main/Tasa%20de%20Abandono%20Interanual%202023-2012%20seg%C3%BAn%20divisi%C3%B3n%20pol%C3%ADtico-territorial.xlsx"

excel_data = pd.ExcelFile(url)
print("Nombres de las hojas:", excel_data.sheet_names)
```

```python

# Obtener el nombre de la primera hoja (índice 0)
nombre_hoja = excel_data.sheet_names[0]
print(nombre_hoja)
```

```python
df = pd.read_excel(url, sheet_name=nombre_hoja)
print(df)
```

```python
df.columns
```

```python
df['Secretaría de Educación de la Nación']
```

```python
# Filtrar ignorando mayúsculas/minúsculas
resultado = df[df['Secretaría de Educación de la Nación'].str.contains("Chaco", case=False, na=False)]
resultado
```

```python
df_copia = df.copy()

# Lista de nuevos nombres de las columnas
columnas_nuevas = [
    'Secretaría de Educación de la Nación', 'Estructura Educativa', 'Total primaria',
    '1° Primaria', '2° Primaria', '3° Primaria', '4° Primaria', '5° Primaria',
    '6° Primaria', '7° Primaria', 'Total Secundaria', '0° Secundaria', '1° Secundaria',
    '2° Secundaria', '3° Secundaria', '4° Secundaria', '5° Secundaria'
]

# Asignar los nuevos nombres de columna a la copia del DataFrame
df_copia.columns = columnas_nuevas

# Mostrar el DataFrame con las columnas renombradas
df_copia
```

```python
resultado_copia = df_copia[df_copia['Secretaría de Educación de la Nación'].str.contains("Chaco", case=False, na=False)]
resultado_copia
```

```python
df_copia.dtypes
```

```python
#Excluir columnas para mostrar datos como el nombre de la provincia
excluir_columnas = ['Secretaría de Educación de la Nación', 'Estructura Educativa']
columnas_a_convertir = df_copia.columns.difference(excluir_columnas)
df_copia[columnas_a_convertir] = df_copia[columnas_a_convertir].apply(pd.to_numeric, errors='coerce')


# Revisar valores NaN
df_copia.isna().sum()
# Llenar los valores NaN con 0 o con un valor adecuado
df_copia = df_copia.fillna(0)  # O con df_copia.fillna(df_copia.mean())
df_copia.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#abandono (por ejemplo, las de Primaria)
df_provincia_primaria = resultado_copia

# Graficar las tasas de abandono para Primaria
df_provincia_primaria.plot(kind='bar', figsize=(10, 6))

plt.title('Tasas de abandono en Chaco')
plt.xlabel('Grado')
plt.ylabel('Tasa de abandono')
plt.xticks(rotation=45)
plt.legend(title='Grado', loc='upper left')
plt.tight_layout()
plt.show()
```
